**Finding the Physiology of 1 monster**

```python
base = 'https://mhworld.kiranico.com/en/monsters'
#get physiology

def get_top_games(html_content):
    
    soup = BeautifulSoup(html_content, 'html.parser')
    table = soup.find("h6", string="Physiology").find_next("table")
    if table:
        #print(table)
        rows = table.find_all("tr") #Physiology table
        #print(rows)
        raw = []
        for row in rows:
            cells = row.find_all("td")
            if cells:
                cell_data = [cell.text.strip() for cell in cells]
                #print(cell_data)
                part = cell_data[0]
                sever = cell_data[1]
                blunt =cell_data[2]
                ranged = cell_data[3]
                fire= cell_data[4]
                water= cell_data[5]
                paralysis= cell_data[6]
                frost= cell_data[7]
                dragon= cell_data[8]
                stun = cell_data[9]
                stamina = cell_data[10]
                raw.append((part,sever,blunt,ranged,fire,water,paralysis,frost,dragon,stun,stamina))
            else:
                #print("cell not found")
                continue
        return pd.DataFrame(raw, columns=('Part', 'Sever','Blunt', 'Ranged', 'Fire', 'Water','Paralysis','Frost', 'Dragon', 'Stun', 'Stamina')).set_index('Part')
    else:
        return pd.DataFrame()

top_games = get_top_games(top_games_html)
top_games
```

**Finding it for All Monsters**

```python
base = 'https://mhworld.kiranico.com/en/monsters'
page_url = 'https://mhworld.kiranico.com/en/monsters'

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.54 Safari/537.36'}
def get_page_html(url):
#    your solution 
    return requests.get(url,headers=headers).content
#    return page texts


top_games_html = get_page_html(page_url)
#print(top_games_html)
soup = BeautifulSoup(top_games_html, 'html.parser')
table = soup.find("tbody")
monsters = table.find_all("tr")
#print(monsters)
monsterURLS = [] # all the urls
for monster in monsters:
    if(monster):
        url = monster.find('a').get("href")
        name = str(url).split('/')[6]
        #print(name)
        monsterURLS.append((name,url))
monsterURLS
```

**Calling To Make Multiple Tables**

```python
monsterframes = {}
for url in monsterURLS:
    #print(url)
    monsterframes[url[0]] = get_top_games(get_page_html(url[1]))
#monsterframes
```
